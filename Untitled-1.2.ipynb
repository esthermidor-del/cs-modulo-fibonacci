```python
# Number of values we want
n = 10
# First two numbers of the series
a = 1
b = 1
# List to store the Fibonacci series
fibonacci = []
for i in range(n):
    fibonacci.append(a)
    a, b = b, a + b

# Show the series
print("Fibonacci series with 10 numbers:")
print(fibonacci)
```
